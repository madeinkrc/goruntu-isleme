```python
def siyah_beyaz(x):
    threshold=150
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            n=x[i,j,0]/3+x[i,j,1]/3+x[i,j,2]/3
            if n>threshold:
                x[i,j]=0
            else:
                x[i,j]=255
```

```python
def reshape(img):
    n = img.shape[0] * img.shape[1]
    img_r = img.reshape(1,n)
    return img_r
```
