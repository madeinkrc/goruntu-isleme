```python
import matplotlib.pyplot as plt 
import numpy as np

img_1=plt.imread("test_2.jpg")
img_1.ndim, img_1.shape
```

```python
img_2=img_1
img_4=np.zeros((img_2.shape))
plt.subplot(1,2,1),plt.imshow(img_2)
plt.subplot(1,2,2),plt.imshow(img_4) #bütün pixelleri 0 olan resim 0:siyah:zeros
plt.show()
```

```python
img_5=np.zeros((img_2.shape[0:2]))
img_5.shape
```

```python
threshold=200
for i in range(img_2.shape[0]): 
    for j in range(img_2.shape[1]): 
        n=img_2[i,j,0]/3+img_2[i,j,1]/3+img_2[i,j,2]/3
        if n>threshold:
            img_5[i,j]=255
        else:
            img_5[i,j]=0     
```

```python
plt.subplot(1,2,1),plt.imshow(img_2)
plt.subplot(1,2,2),plt.imshow(img_5, plt.cm.binary)
plt.show()
```
