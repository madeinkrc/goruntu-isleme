```python
img_3=np.zeros((img_2.shape[0:2]))
img_3.shape
```

```python
img_4=np.zeros((img_2.shape[0:2]))
img_4.shape
```

```python
threshold=100
for i in range(img_2.shape[0]): #384
    for j in range(img_2.shape[1]): #512
        n=img_2[i,j,0]/3+img_2[i,j,1]/3+img_2[i,j,2]/3
        img_3[i,j]=n
        if n>threshold:
            img_4[i,j]=255
        else:
            img_4[i,j]=0   
```

```python
#BLACK-WHITE
plt.imshow(img_4, plt.cm.binary)
plt.show()
```
