```python
import math
def findDistance(x,y):
    return math.sqrt(((x[0]-y[0])**2)+((x[1]-y[1])**2))
```

```python
from scipy import misc
import matplotlib.pyplot as plt
f = misc.face()
plt.imshow(f)
plt.show()
```

```python
im_size=f.shape
im_size
```

```python
center=[im_size[0]/2,im_size[1]/2]
center
```

```python
#Her i,j koordinatının merkeze olan uzaklığı eğer 350 pixelden fazlaysa onların değerini 255 (beyaz) yap.
for i in range(im_size[0]):
    for j in range(im_size[1]):
        if (findDistance([i,j],center)>350):
            f[i,j,:]=255,255,255
```

```python
plt.imshow(f)
plt.show()
```
