```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
#ilgili resim c klasöründe!
img=mpimg.imread(r'c:\stinkbug.png')
```

```python
#3 RGB ve Boyutları.
img.ndim,img.shape
```

```python
#img ile img_20 birbirinin transpozu değerler olmalı.
#zeros fonksiyonu pixel değeri 0 olan dizi oluşturur zeros:0:siyah - ones:1:beyaz
img_20=np.zeros((500,375,3)) 
img_20.shape
```

```python
plt.imshow(img_20)
plt.show()
```

```python
#Bu döngü resimdeki bütün rgb değerlerini veriyor.
for i in range(375):
    for j in range(500):
        #transpoze işlemi -x,y ye göre-
        img_20[j,i,:]=img[i,j,:]
        #print(img[i,j,:])
```

```python
plt.imshow(img_20)
plt.show()
```

```python
img_30=np.zeros((500,375,3)) 
img_30.shape
for i in range(375): #500
    for j in range(500): #375
        #resmin negatifini alıyor.
        img_30[j,i,:]=1-img[i,j,:] #i,j - img_20
```

```python
plt.imshow(img_30)
plt.show()
```

```python
img_40=np.zeros((375,500,3)) 
img_40.shape
for i in range(375):
    for j in range(500):
        #x ve y ye göre simetriğini aldı. yani orjine göre ayna görüntüsünü aldı.
        #500-j-1 yerine j yazsaydık sadece x eksenine göre simetrisini alacaktı.
        img_40[375-i-1,500-j-1,:]=img[i,j,:]
```

```python
plt.imshow(img_40)
plt.show()
```

```python
plt.subplot(1,4,1),plt.imshow(img)
plt.subplot(1,4,2),plt.imshow(img_20)
plt.subplot(1,4,3),plt.imshow(img_30)
plt.subplot(1,4,4),plt.imshow(img_40)
plt.show()
```
