```python
img_5=np.zeros((img_1.shape[0:2]))
img_5.shape
```

```python
threshold=200
for i in range(img_1.shape[0]): 
    for j in range(img_1.shape[1]): 
        n=img_1[i,j,0]/3+img_1[i,j,1]/3+img_1[i,j,2]/3
        if n>threshold:
            img_5[i,j]=255
        else:
            img_5[i,j]=0  
```

```python
def get_Histogram(image):
    my_Histogram={}
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            item=image[i,j]
            if item in my_Histogram.keys():
                my_Histogram[item]=my_Histogram[item]+1
            else:
                my_Histogram[item]=1
    return my_Histogram
```

```python
my_Histogram = get_Histogram(img_5)
my_Histogram
```

```python
x=my_Histogram[0.0]
y=my_Histogram[255.0]
print ("Siyah Beyaz Orani :",y/x)
```
